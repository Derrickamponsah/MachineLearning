```python
import numpy as np
from matplotlib import pyplot as plt
import sklearn
```

```python
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
```

```python
data
```

```python
print(data.DESCR)
```

```python
data.data[0]
```

```python
data.target[0]
```

```python
x_input = data['data']
```

```python
x_input
```

```python
y_output = data['target']
```

```python
y_output
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_input, y_output, test_size = 0.2)
```

```python
x_train.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted[: 20]
```

```python
y_test[: 20]
```

```python
plt.plot(y_train[:80], y_test[:80], 'o')
plt.plot(y_predicted[:80], y_test[:80], 'o')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_predicted, y_test)
```

```python
mse
```

```python
r2_scored = r2_score(y_predicted, y_test)
```

```python
r2_scored
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF_model = RandomForestRegressor()
```

```python
RF_model.fit(x_train, y_train)
```

```python
RF_predicted = RF_model.predict(x_test)
```

```python
RF_predicted[:20]
```

```python
y_test[: 20]
```

```python
RF_mse = mean_squared_error(RF_predicted, y_test)
```

```python
RF_mse
```

```python
RF_r2 = r2_score(RF_predicted, y_test)
```

```python
RF_r2
```

```python
from sklearn.svm import SVR
```

```python
svm_model = SVR()
```

```python
svm_model.fit(x_train, y_train)
```

```python

svm_predicted = svm_model.predict(x_test)
```

```python
svm_predicted[: 20]
```

```python
y_test[: 20]
```

```python
svm_mse = mean_squared_error(svm_predicted, y_test)
```

```python
svm_mse
```

```python
svm_scored = r2_score(svm_predicted, y_test)
```

```python
svm_scored
```

```python
results = {'svm_mse': svm_mse,
            'svm_scored': svm_scored,}
```

```python
import pandas as pd
```

```python
results
```

```python
print(f'SVM MSE: {svm_mse}')
print(f'SVM RS: {svm_scored}')
print(f'RF MSE: {RF_mse}')
print(f'RF RS: {RF_r2}')
print(f'DT MSE: {mse}')
print(f'DT RS: {r2_scored}')
```
