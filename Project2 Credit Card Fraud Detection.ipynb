```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card = pd.read_csv('creditcard.csv')
```

```python
# distribution of legit transaction and fraudulent transaction
credit_card['Class'].value_counts()
```

DATASET IS HIGHLY IMBALANCE

0 --> Normal Transaction
1 --> Fraudulent Transaction

```python
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Compare values for both transaction
credit_card.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing same distribution of normal and fraudulent
transaction

Number of frauduent transaction --> 492

```python
legit_sample = legit.sample(n = 492)
```

Concatenating two dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting data into features and targets

```python
X = new_dataset.drop(columns = 'Class' , axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Using Logistic Regression

```python
# Training the logistic regression model with training data
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_predicted = model.predict(X_train)
training_accuracy = accuracy_score(Y_train, X_train_predicted)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy on Training data:', training_accuracy)
```

```python
# accuracy on testing data
X_test_predicted = model.predict(X_test)
testing_accuracy = accuracy_score(Y_test, X_test_predicted)
```

```python
print('Accuracy on Testing data:', testing_accuracy)
```
